# The Monty Hall Problem

The Monty Hall problem is a counter-intuitive statistics puzzle:
- There are 3 doors, behind which are two goats and a car.
- You pick a door (call it door 1). Youâ€™re hoping for the car of course.
- Monty Hall, the game show host, examines the other doors (2 & 3) and opens one with a goat. (If both doors have goats, he picks randomly.)

There are actually a number of assumptions:
- Monty will always open a door.
- Monty never opens the door you have chosen.
- Monty never opens the door with the car behind it.
- The car is equally likely to be behind any door.
- Given a choice of doors, Monty chooses at random.

# Monty Hall Problem 10 million trials:

```python
import random

trials = 10000000

stay = []
switch = []
for i in range(trials):
    doors = ['car', 'goat', 'goat']
    random.shuffle(doors)
    choice = random.choice([0, 1, 2])
    stay_result = doors.pop(choice)
    doors.remove('goat')
    switch_result = doors[0]
    stay.append(stay_result)
    switch.append(switch_result)
    
print('If you choose to switch, the probability of winning the car is:', stay.count('goat')/trials)
print('If you choose not to switch, the probability of winning the car is:', switch.count('goat')/trials)
```
